```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
```

```python
data = xr.open_dataset('CERES_EBAF_Ed4.1_Subset_200101-200102.nc')
```

```python
print(data.variables)
lat = np.array(data['lat'])
lon = np.array(data['lon'])
lonGrid, latGrid = np.meshgrid(lon, lat)
```

```python
fig = plt.figure(figsize=(12,6))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.grid(True)

c1 = ax.contourf(lonGrid, latGrid, data['sfc_sw_up_all_mon'][0], cmap='Spectral_r')
fig.colorbar(c1,label=(data['sfc_sw_up_all_mon'].long_name +' ('+ data['sfc_sw_up_all_mon'].units+')'), orientation='horizontal')
ax.coastlines()

ax.set_title('January ' + data['sfc_sw_up_all_mon'].long_name)
```
